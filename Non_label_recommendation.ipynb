```python
import pickle
import numpy as np
import pandas as pd
```

```python
# 데이터 불러오기
data = pd.read_csv('sk_data.csv')
job_wage = pd.read_csv('job_wage.csv')
with open('recommend_dict', 'rb') as fr:
     recommend_dict = pickle.load(fr)
```

```python
## 연봉이 낮으면 낮은 숫자
job_dict = {}

for num,i in enumerate(job_wage[::-1]['직업명']):
    job_dict[i] = num+1
```

```python
job_wage
```

```python
job_dict['학생'] = 0
```

```python
data['직업']=data['직업'].replace(job_dict)
```

```python
del data['Unnamed: 0']
```

```python
data[:12]
```

```python
del data['person_id']
```

```python
del data['주식']
del data['선물옵션']
del data['장내채권']
del data['월']
```

```python
data
```

```python
def mean_norm(df_input):
    return df_input.apply(lambda x: (x-x.mean())/ x.std(), axis=0)
```

```python
data['월'] = data['월']/12
data['나이'] = data['나이']/4
data['직업'] = data['직업']/622
data[['소득(연)','자산','월별 카드 값']] = mean_norm(data.loc[:,['소득(연)','자산','월별 카드 값']])
```

```python
re_df=data[:96000]
no_df=data[96000:]
```

```python
re_df
```

```python
re_array=re_df.to_numpy()
```

```python
re_array.shape
```

```python
re_array=re_array.reshape(8000,12,7)
```

```python
re_array
```

```python
cha_dict = {'증권사1':[1,0,0],'증권사2':[0,1,0],'증권사3':[0,0,1]}
y_list = []
for i in recommend_dict.values():
    y_list.append(cha_dict[i])
```

```python
y=np.array(y_list)
```

```python
re_array.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(re_array,y,test_size=0.2,shuffle=True,random_state=1004)
```

# 1D CNN classification model

```python
import tensorflow as tf
import keras

from keras import regularizers
from keras.layers import Flatten
from sklearn.model_selection import train_test_split
import tensorflow.keras.backend as K
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input,Conv1D,Dense,Dropout,MaxPooling1D,ZeroPadding1D, Flatten,BatchNormalization,Lambda

import matplotlib.pyplot as plt
```

```python
inputShape = (12,7)
model = Sequential()
model.add(Conv1D(16, 3, padding="same",input_shape = inputShape, activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling1D(2))
model.add(Dropout(0.3))

model.add(Conv1D(32, 3, padding="same",input_shape = inputShape, activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling1D(2))
model.add(Dropout(0.3))

model.add(Conv1D(32, 3, padding="same",input_shape = inputShape, activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling1D(2))
model.add(Dropout(0.3))

model.add(Flatten())
model.add(Dense(3,activation = 'softmax'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    batch_size=32,
    epochs=50)
```

```python
final_predict=model.predict(np.array(no_df).reshape(2000,12,7))
```

```python
no_recommend_dict = {}
for i in range(2000):
  no_recommend_dict[i+18000]='증권사'+str(np.argmin(final_predict[i])+1)
```

```python
recommend_dict.update(no_recommend_dict)
recommend_dict
```
