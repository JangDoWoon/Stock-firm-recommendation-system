```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('sk_data.csv')
```

```python
del data['Unnamed: 0']
```

```python
data
```

```python
df = data.iloc[:96000]
```

```python
1000 * 0.005
```

```python
# 증권사 수수료, 혜택
ju_dict = {}
su_dict = {}
cha_dict = {}
he_dict = {}
ch_list = ['증권사'+str(i) for i in range(1,4)]
j = 0
ju_dict[ch_list[0]] = 0.01
ju_dict[ch_list[1]] = 0.015
ju_dict[ch_list[2]] = 0.01

su_dict[ch_list[0]] = 0.015
su_dict[ch_list[1]] = 0.01
su_dict[ch_list[2]] = 0.01

cha_dict[ch_list[0]] = 0.01
cha_dict[ch_list[1]] = 0.01
cha_dict[ch_list[2]] = 0.015

he_dict[ch_list[0]] = 1
he_dict[ch_list[1]] = 3.5
he_dict[ch_list[2]] = 1
```

```python
df
```

```python
person_id = [i for i in range(10000,18000)]
```

```python
# 각 사람별로 구매한 주식, 선물옵션, 장내채권을 넣었음
sum_ju_dict = {}
sum_su_dict = {}
sum_cha_dict = {}
for person in person_id:
    sum_ju_dict[person]=df[df['person_id'] == person].sum()['주식']
    sum_su_dict[person]=df[df['person_id'] == person].sum()['선물옵션']
    sum_cha_dict[person]=df[df['person_id'] == person].sum()['장내채권']
```

```python
# 각 사람별 수수료 계산 후 가장 작은 수수료가 있는 증권사를 선택함 
recommend_dict = {}
for person in person_id:
    final_dict = {}
    for chapital in ch_list:
        final_dict[chapital] = sum_ju_dict[person]*ju_dict[chapital] + sum_su_dict[person]*su_dict[chapital] + sum_cha_dict[person]*cha_dict[chapital] - he_dict[chapital]
    recommend_dict[person] = min(final_dict, key= lambda x : final_dict[x])
```

```python
recommend_dict
```

```python
person = int(input("person_id를 입력하세요: "))
print(str(person)+'의 추천 증권사는 '+str(recommend_dict[person])+' 입니다.')
```

```python
import pickle

with open('recommend_dict','wb') as fw:
    pickle.dump(recommend_dict, fw)
```

```python
with open('recommend_dict', 'rb') as fr:
    user_loaded = pickle.load(fr)
```
